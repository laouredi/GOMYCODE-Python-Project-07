```python
import pandas as pd
import numpy as np

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data, index=labels)

# 1 - Print the three first rows using the head() method.
print("The first three rows of the DataFrame:")
print(df.head(3))

# 2 - Delete rows with NaN values.
df = df.dropna()
print("\nDataFrame after removing rows with NaN values:")
print(df)

# 3 - Extract the 'name' and 'score' columns from the DataFrame.
df = df[['name', 'score']]
print("\nDataFrame with only the 'name' and 'score' columns:")
print(df)

# 4 - Write a Pandas program to append a new row 'k' to the DataFrame with these values (name: "Suresh", score: 15.5, attempts: 1, qualify: "yes").
new_row = pd.DataFrame({'name': ['Suresh'], 'score': [15.5], 'attempts': [1], 'qualify': ['yes']}, index=['k'])
df = df.append(new_row)
print("\nDataFrame after adding a new row:")
print(df)

# 5 - Write a Pandas program to delete the 'attempts' column from the DataFrame.
df = df.drop(['attempts'], axis=1)
print("\nDataFrame after removing the 'attempts' column:")
print(df)

# 6 - Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0.
df['Success'] = np.where(df['score'] > 10, 1, 0)
print("\nDataFrame after adding a new 'Success' column:")
print(df)

# 7 - After executing the final DataFrame, export it into a CSV file named "my_data".
df.to_csv("my_data.csv", index=False)
```
